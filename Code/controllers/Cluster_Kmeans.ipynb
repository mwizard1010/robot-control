```python
from sklearn.cluster import KMeans
import pickle
resultsFolder = 'results/tests/'
with open(resultsFolder + 'agentRL4.modelstate', "rb") as f:
    states = pickle.load(f)
# print(states[0])
```

```python
import math
states_array = [x for x in states if math.isnan(x[0]) == False and math.isnan(x[1]) == False]
```

```python
import matplotlib.pyplot as plt
Sum_of_squared_distances = []
K = range(1,10)
print()
for num_clusters in K :
    print(num_clusters)
    kmeans = KMeans(n_clusters=num_clusters)
    kmeans.fit(states_array)
    Sum_of_squared_distances.append(kmeans.inertia_)

plt.plot(K,Sum_of_squared_distances,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Sum of squared distances/Inertia') 
plt.title('Elbow Method For Optimal k')
plt.savefig('elbow.pdf')
plt.show()
```
